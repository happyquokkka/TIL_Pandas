```python
# 각 요일별 테이블당 인원수의 평균과 최대인원 최소인원을 피봇테이블로 확인하시오
tips.pivot_table('size', index='day', aggfunc=['mean', 'max', 'min'])
```

```python
# 각 요일별 테이블당 인원수의 평균과 최대인원 최소인원을 피봇테이블로 확인하시오
tips.pivot_table('size', index=['day','time'], aggfunc=['mean','max','min'])
```

```python
# 각 요일별 테이블당 인원수의 평균과 최대인원 최소인원을 피봇테이블로 확인하시오
tips.pivot_table('size',index='time',columns='day',aggfunc=['mean','max','min'])
```

```python
# 각 요일에 따른 time별 tip_pt의 평균을 확인하시오
pd.DataFrame(tips.groupby(['day','time'])['tip_pt'].mean())
```

#### 연습문제 3

- 타이타닉 승객 데이터를 이용하여 다음 분석을 실시하라. 데이터는 다음과 같이 받을 수 있다.
    - titanic = sns.load_dataset("titanic")
    
- 문제 1

- qcut 명령으로 세 개의 나이 그룹['청년','중년','장년']을 만든다.



- 성별, 선실, 나이 그룹에 의한 생존율을 데이터프레임으로 계산한다. 
    - 행에는 성별 및 나이 그룹에 대한 다중 인덱스를 사용하고 열에는 선실 인덱스를 사용한다. 
    - 생존률은 해당 그룹의 생존 인원수를 전체 인원수로 나눈 값이다.
    

- 성별 및 선실에 의한 생존율을 피봇 데이터 형태로 만든다.

```python
# 데이터 로드하기
import seaborn as sns
titanic = sns.load_dataset("titanic")
```

```python
# qcut 명령으로 세 개의 나이 그룹['청년','중년','장년']의 범주형 값을 만들어  age_group 열을 생성한다
agelist = titanic['age']
agelist
titanic['age_group'] = pd.qcut(agelist, 3, labels=['청년', '중년', '장년'])
```

```python
titanic.head()
```

- 성별, 선실, 나이 그룹에 의한 생존율을  계산한다. 

    - 행에는 성별로 나이그룹을 구분하는 다중 인덱스를 정의하고 열에는 class별 생존구분을 표현하는 pivot 테이블을 생성한다 

```python
# 피봇 테이블 생성 
#피봇 테이블 생성코드
titanic['alive']
titanic_df = titanic.pivot_table(values='survived', index=['sex', 'age_group'], columns=['class', 'alive'], aggfunc=['count']) 
```

```python
titanic_df
```

```python
# First 선실의 성별로 age_group의 인원 계산 코드
# First 선실의 청년 중년 장년의 각 인원수는 22,19,44 명임

titanic_df[('count','First')].sum(axis=1)
type(titanic_df[('count','First')].sum(axis=1))
```

```python
# 생존률은 해당 그룹의 생존 인원수를 전체 인원수로 나눈 값이다.
# 생존율 계산 후 데이터 프레임 생성
titanic_df

# 선실 등급별 전체 인원수
total_first = titanic_df[('count','First')].sum(axis=1)
total_second = titanic_df[('count','Second')].sum(axis=1)
total_third = titanic_df[('count','Third')].sum(axis=1)

# 선실 등급별 생존 인원수
alive_first = titanic_df[('count','First','yes')]
alive_second = titanic_df[('count','Second','yes')]
alive_third = titanic_df[('count','Third','yes')]

data= {'First': alive_first/total_first*100,  #First 생존율 계산,
       'Second': alive_second/total_second*100, #Second 생존율 계산,
       'Third' : alive_third/total_third*100  #Third 생존율 계산
      }




titanic_survived_rate = pd.DataFrame(data,columns=['First','Second','Third'])
titanic_survived_rate
```

```python
# titanic_survived_rate 데이터프레임을 사용해 성별 각 선실의 생존율을 피봇 테이블로 생성
pd.pivot_table(titanic_survived_rate, index='sex')
```
